```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime
```

```python
df = pd.read_csv('train_data.csv')
```

```python
df = df.rename(columns={'Total Stops': 'Total_Stops'})
```

```python
df = df.dropna()
```

```python
import pandas as pd
from datetime import datetime
 
def is_weekend(date_str):
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")

    day_of_week = date_obj.weekday()
 
    if day_of_week >= 4:
        return 1
    else:
        return 0

df['Weekend'] = df['StandardDateofJourney'].apply(is_weekend)
```

```python
df["Dep_Hour"] = pd.to_datetime(df["Departure_Time"]).dt.hour
```

```python
df["Arrival_Hour"] = pd.to_datetime(df["Arrival_Time"]).dt.hour
```

```python
df['Total_Stops'] = df['Total_Stops'].apply(lambda x: '1-stop' if x.strip().startswith('1-stop') else x)

print(df['Total_Stops'].value_counts())
```

```python
df['StandardDateofBooking'] = pd.to_datetime(df['StandardDateofBooking'], format='%Y-%m-%d')
df['StandardDateofJourney'] = pd.to_datetime(df['StandardDateofJourney'], format='%Y-%m-%d')

df['Difference_in_days'] = (df['StandardDateofJourney'] - df['StandardDateofBooking']).dt.days

df.sample(20)
```

```python
import pandas as pd

negative_count = (df['Difference_in_days'] < 0).sum()

print(f"Number of negative values in 'Difference_in_days': {negative_count}")
```

```python
df = df[df['Difference_in_days'] >= 0]

print("DataFrame after removing negative values:")

df.info()
```

```python
def flight_dep_time(hour):
    if hour == -1:
        return "missing"
    elif (hour > 4) and (hour <= 8):
        return "Early Morning"
    elif (hour > 8) and (hour <= 12):
        return "Morning"
    elif (hour > 12) and (hour <= 16):
        return "Noon"
    elif (hour > 16) and (hour <= 20):
        return "Evening"
    elif (hour > 20) and (hour <= 24):
        return "Night"
    else:
        return "Late Night" 

df['Departure_Time_Category'] = df['Dep_Hour'].apply(flight_dep_time)
df['Arrival_Time_Category'] = df['Arrival_Hour'].apply(flight_dep_time)
```

```python
df['Journey_Day'] = df['StandardDateofJourney'].dt.day
```

```python
df = df.drop(['StandardDateofBooking', 'StandardDateofJourney', 'Flight_Number', 'Departure_Time', 'Arrival_Time', 'Duration_in_mins', 'Dep_Hour', 'Arrival_Hour'], axis=1)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Define the mapping for Flight_Class
flight_class_mapping = {
    'FIRST': 4,
    'BUSINESS': 3,
    'PREMIUMECONOMY': 2,
    'ECONOMY': 1
}

# Define the mapping for Total_Stops
total_stops_mapping = {
    '2+-stops': 3,
    '1-stop': 2,
    'non-stop': 1
}
```

```python
# Apply the mapping to the Flight_Class column
df['Flight_Class'] = df['Flight_Class'].map(flight_class_mapping)

# Apply the mapping to the Total_Stops column
df['Total_Stops'] = df['Total_Stops'].map(total_stops_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder

# Define the categorical columns to encode
categorical_columns = ['Airline_Company', 'Departure_City', 'Arrival_City', 'Departure_Time_Category', 'Arrival_Time_Category']

# Initialize a dictionary to store the LabelEncoders
label_encoders = {}

# Apply LabelEncoder to the specified columns
for column in categorical_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
df = df.dropna()
```

```python
df.to_csv('transformed_train_data.csv', index=False)
```

```python
train_data = pd.read_csv('transformed_train_data.csv')

X_train = train_data.drop(['Price'], axis=1)
y_train = train_data['Price']

test_data = pd.read_csv('transformed_test_data.csv')
X_test = test_data.drop(['Price'], axis=1)
y_test = test_data['Price']
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pickle

# Create the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)

# Calculate RMSE
rmse = np.sqrt(mse)

# Calculate R2 Score
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R2 Score: {r2}')
```

```python
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
dt_y_pred = dt_model.predict(X_test)
print('Decision Tree:')
print(f'MAE: {mean_absolute_error(y_test, dt_y_pred)}')
print(f'MSE: {mean_squared_error(y_test, dt_y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, dt_y_pred))}')
print(f'R2 Score: {r2_score(y_test, dt_y_pred)}')
print("<------------------------------------------>")

rf_model = RandomForestRegressor(random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)
rf_y_pred = rf_model.predict(X_test)
print('Random Forest:')
print(f'MAE: {mean_absolute_error(y_test, rf_y_pred)}')
print(f'MSE: {mean_squared_error(y_test, rf_y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, rf_y_pred))}')
print(f'R2 Score: {r2_score(y_test, rf_y_pred)}')
print("<------------------------------------------>")

gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train, y_train)
gb_y_pred = gb_model.predict(X_test)
print('Gradient Boosting:')
print(f'MAE: {mean_absolute_error(y_test, gb_y_pred)}')
print(f'MSE: {mean_squared_error(y_test, gb_y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, gb_y_pred))}')
print(f'R2 Score: {r2_score(y_test, gb_y_pred)}')
print("<------------------------------------------>")

xgb_model = XGBRegressor(random_state=42, n_jobs=-1)
xgb_model.fit(X_train, y_train)
xgb_y_pred = xgb_model.predict(X_test)
print('XGBoost:')
print(f'MAE: {mean_absolute_error(y_test, xgb_y_pred)}')
print(f'MSE: {mean_squared_error(y_test, xgb_y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, xgb_y_pred))}')
print(f'R2 Score: {r2_score(y_test, xgb_y_pred)}')
```
