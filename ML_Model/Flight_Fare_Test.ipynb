```python
import pandas as pd
from datetime import datetime
 
def is_weekend(date_str):
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")

    day_of_week = date_obj.weekday()
 
    if day_of_week >= 4:
        return 1
    else:
        return 0

df['Weekend'] = df['StandardDateofJourney'].apply(is_weekend)

df.sample(20)
```

```python
df["Dep_Hour"] = pd.to_datetime(df["Departure_Time"]).dt.hour
```

```python
df["Arrival_Hour"] = pd.to_datetime(df["Arrival_Time"]).dt.hour
```

```python
df.sample(5)
```

```python
df['Total_Stops'] = df['Total_Stops'].apply(lambda x: '1-stop' if x.strip().startswith('1-stop') else x)

print(df['Total_Stops'].value_counts())
```

```python
df['StandardDateofBooking'] = pd.to_datetime(df['StandardDateofBooking'], format='%Y-%m-%d')
df['StandardDateofJourney'] = pd.to_datetime(df['StandardDateofJourney'], format='%Y-%m-%d')

df['Difference_in_days'] = (df['StandardDateofJourney'] - df['StandardDateofBooking']).dt.days

df.sample(20)
```

```python
import pandas as pd

negative_count = (df['Difference_in_days'] < 0).sum()

print(f"Number of negative values in 'Difference_in_days': {negative_count}")
```

```python
df = df[df['Difference_in_days'] >= 0]

print("DataFrame after removing negative values:")

df.info()
```

```python
def flight_dep_time(hour):
    if hour == -1:
        return "missing"
    elif (hour > 4) and (hour <= 8):
        return "Early Morning"
    elif (hour > 8) and (hour <= 12):
        return "Morning"
    elif (hour > 12) and (hour <= 16):
        return "Noon"
    elif (hour > 16) and (hour <= 20):
        return "Evening"
    elif (hour > 20) and (hour <= 24):
        return "Night"
    else:
        return "Late Night" 

df['Departure_Time_Category'] = df['Dep_Hour'].apply(flight_dep_time)
df['Arrival_Time_Category'] = df['Arrival_Hour'].apply(flight_dep_time)
```

```python
df['Journey_Day'] = df['StandardDateofJourney'].dt.day
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Define the mapping for Flight_Class
flight_class_mapping = {
    'FIRST': 4,
    'BUSINESS': 3,
    'PREMIUMECONOMY': 2,
    'ECONOMY': 1
}

# Define the mapping for Total_Stops
total_stops_mapping = {
    '2+-stops': 3,
    '1-stop': 2,
    'non-stop': 1
}
```

```python
# Apply the mapping to the Flight_Class column
df['Flight_Class'] = df['Flight_Class'].map(flight_class_mapping)

# Apply the mapping to the Total_Stops column
df['Total_Stops'] = df['Total_Stops'].map(total_stops_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder

# Define the categorical columns to encode
categorical_columns = ['Airline_Company', 'Departure_City', 'Arrival_City', 'Departure_Time_Category', 'Arrival_Time_Category']

# Initialize a dictionary to store the LabelEncoders
label_encoders = {}

# Apply LabelEncoder to the specified columns
for column in categorical_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
df.to_csv('transformed_test_data.csv', index=False)
```

```python
import pickle

# Save each LabelEncoder as a .pkl file
for column, le in label_encoders.items():
    with open(f'label_encoder_{column}.pkl', 'wb') as file:
        pickle.dump(le, file)

# Save the manually encoded mappings
with open('flight_class_mapping.pkl', 'wb') as file:
    pickle.dump(flight_class_mapping, file)

with open('total_stops_mapping.pkl', 'wb') as file:
    pickle.dump(total_stops_mapping, file)
```
